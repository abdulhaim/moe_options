# Moe Option Critic Model Versus Naive Option Critic  

```python
import warnings; warnings.simplefilter('ignore')

import seaborn as sns
import pylab as plot
import matplotlib.pyplot as plt
import numpy as np
import os
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator

import numpy as np
import matplotlib.pyplot as plt
from statistics import mean
sns.set_style("ticks")
sns.set_palette(sns.color_palette("hls", 8))
plt.rc('text', usetex=False)
plt.rc('font', family='serif')
params = {'legend.fontsize': 14}
plot.rcParams.update(params)

MAX_ITERATION = 20
```

# MoE Options Model 

```python
d = '/Users/Marwa/Documents/IAP/Research/option-critic-pytorch-moe/runs'
folder_runs = [os.path.join(d, o) for o in os.listdir(d) 
                    if os.path.isdir(os.path.join(d,o))]

y_axis = []

for file in folder_runs:
    event_acc = EventAccumulator(file)
    event_acc.Reload()
    w_times, step_nums, vals = zip(*event_acc.Scalars('episode_lengths'))
    y_axis.append(list(vals))


y_axis_moe = list(map(list, zip(*y_axis))) 

mean_moe = np.mean(y_axis_moe[0:800], axis=1)
std_moe = np.std(y_axis_moe[0:800], axis=1)
confidence_moe = 1.96 * std_moe / float(np.sqrt(20))  # 95% interval
error_moe = (mean_moe - confidence_moe, mean_moe + confidence_moe)
```

# Naive Options Model

```python
d = '/Users/Marwa/Documents/IAP/Research/option-critic-pytorch-naive/runs'
folder_runs = [os.path.join(d, o) for o in os.listdir(d) 
                    if os.path.isdir(os.path.join(d,o))]

y_axis = []

for file in folder_runs:
    event_acc = EventAccumulator(file)
    event_acc.Reload()
    w_times, step_nums, vals = zip(*event_acc.Scalars('episode_lengths'))
    y_axis.append(list(vals))

y_axis_naive = list(map(list, zip(*y_axis))) 

mean_naive = np.mean(y_axis_naive, axis=1)
std_naive = np.std(y_axis_naive, axis=1)
confidence_naive = 1.96 * std_naive / float(np.sqrt(10))  # 95% interval
error_naive = (mean_naive - confidence_naive, mean_naive + confidence_naive)
```

```python
fig, ax = plt.subplots()
sns.set_style("ticks")

legend_moe = 'moe options'  
ax.fill_between(np.arange(mean_moe.shape[0]), error_moe[0], error_moe[1], alpha=0.25, facecolor = "yellow")
ax.plot(np.arange(mean_moe.shape[0]), mean_moe, markersize=0.01,marker='o', label=legend_moe)  

legend_naive = 'naive options'  
ax.fill_between(np.arange(mean_naive.shape[0]), error_naive[0], error_naive[1], alpha=0.25, facecolor = "red")
ax.plot(np.arange(mean_naive.shape[0]), mean_naive, markersize=0.01,marker='o',label=legend_naive)  

plt.xlabel('Episodes', size=16)
plt.ylabel('Steps', size=16)

legend = plt.legend(
    frameon=False,
    framealpha=0.2,
    bbox_to_anchor=(0., 1.03, 1., .102), 
    loc=2,
    ncol=2, 
    mode="expand", 
    borderaxespad=0.)
```
